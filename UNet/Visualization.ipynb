```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import os
import matplotlib.pyplot as plt
```

```python
# Define the dataset path
train_folder = '/content/drive/MyDrive/Data_sets/PETR/Spill_Data/train'
```

```python
# Define the directories containing the images
no_spill_dir = "/content/drive/MyDrive/Data_sets/PETR/Spill_Data/Train/NoSpill/"
oil_spill_dir = "/content/drive/MyDrive/Data_sets/PETR/Spill_Data/Train/Oilspill/"

# Get a list of the filenames in each directory
no_spill_files = os.listdir(no_spill_dir)[:5]
oil_spill_files = os.listdir(oil_spill_dir)[:5]
```

```python
# Load and resize the images from the NoSpill directory
no_spill_images = []
for filename in no_spill_files:
    filepath = os.path.join(no_spill_dir, filename)
    img = cv2.imread(filepath)
    img_resized = cv2.resize(img, (300, 300))
    no_spill_images.append(img_resized)

# Load and resize the images from the Oilspill directory
oil_spill_images = []
for filename in oil_spill_files:
    filepath = os.path.join(oil_spill_dir, filename)
    img = cv2.imread(filepath)
    img_resized = cv2.resize(img, (300, 300))
    oil_spill_images.append(img_resized)
```

```python
# Concatenate the NoSpill images and Oilspill images horizontally with white space in between
print("\n")
print("\t\t\t\t{:^40}".format("No Spill Images"))
white_space = np.full((300, 10, 3), 255, dtype=np.uint8)
no_spill_concatenated = np.concatenate(no_spill_images, axis=1)
final_image = np.concatenate((no_spill_concatenated, white_space), axis=1)
cv2_imshow(final_image)
```

```python
print("\n")
print("\t\t\t\t{:^40}".format("Oil Spill Images"))
oil_spill_concatenated = np.concatenate(oil_spill_images, axis=1)
final_image = np.concatenate((oil_spill_concatenated, white_space), axis=1)

# Display the final image
cv2_imshow(final_image)
```

```python
no_spill_count = len(os.listdir(no_spill_dir))
oil_spill_count = len(os.listdir(oil_spill_dir))

plt.bar(["No Spill", "Oil Spill"], [no_spill_count, oil_spill_count])
plt.ylim(top=180)
plt.title('Number of Images in Train Dataset')
for i, v in enumerate([no_spill_count, oil_spill_count]):
    plt.text(i, v + 5, str(v), ha="center")

plt.show()
```
